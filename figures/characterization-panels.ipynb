#### Load full end-to-end models

```python
for folder in list(np.sort(glob.glob("stats-panel/full-model*"))):
    checkpoint_path = f"{folder}/lightning_logs/version_0/checkpoints/epoch=4-step=25015.ckpt"
    model = EndToEndConvNN_PL.load_from_checkpoint(checkpoint_path)
    model.eval()
    model.freeze()
    model.cpu()
    rand_seed = int(folder.split("_")[1].split("-")[0])
    lambda_RC = int(folder.split("_")[-1])
    models = pd.concat([models, pd.DataFrame({"rand_seed": [rand_seed], "lambda_RC": [lambda_RC], "model_type":"full", "model": [model]})]).reset_index(drop=True)
```

```python
valid_data = dict()
for rand_seed in models["rand_seed"].unique():
    L.seed_everything(rand_seed)
    _, valid_dset = torch.utils.data.random_split(dset, [0.75, 0.25])  # type: ignore
    valid_data[rand_seed] = DataLoader(valid_dset, batch_size=batch_size, shuffle=False, drop_last=True, num_workers=12)
```

## Summary plots

```python
# models_with_r2 = pd.read_csv("models_with_r2.csv")
models_with_r2 = pd.read_csv("models_with_r2_estimates.csv")
feat_names = [x.split("r2_score_")[1] for x in models_with_r2.columns if "r2_score_" in x]
models_with_r2["num_eods"] = 1
# models_with_r2_multiple_eods = pd.read_csv("models_with_r2_multiple_eods.csv")
models_with_r2_multiple_eods = pd.read_csv("models_with_r2_multiple_eods_estimates.csv")
models_summary = models_with_r2.copy()
num_eods_simulated = [int(x.split("num_eods_")[1]) for x in models_with_r2_multiple_eods.columns if "r2_score_" in x]
for num_eods in num_eods_simulated:
    temp_models = models_with_r2_multiple_eods[["rand_seed", "lambda_RC", "model_type"] + [f"r2_score_{feat_name}_num_eods_{num_eods}" for feat_name in feat_names]].copy()
    for feat_name in feat_names:
        temp_models[f"r2_score_{feat_name}"] = temp_models[f"r2_score_{feat_name}_num_eods_{num_eods}"]
        temp_models.drop(columns=[f"r2_score_{feat_name}_num_eods_{num_eods}"], inplace=True)
    temp_models["num_eods"] = num_eods
    models_summary = pd.concat([models_summary, temp_models]).reset_index(drop=True)
```

# Performance as number of EODs

```python
results_estimates = pd.read_csv("models_with_r2_multiple_eods_estimates.csv")
results_estimates.loc[results_estimates["model_type"] == "full", "model_type"] = "no feedback"
results_values = pd.read_csv("models_with_r2_multiple_eods.csv")
results_values.loc[results_values["model_type"] == "feedback", [x for x in results_values.columns if "radii" in x]] = results_values.loc[results_values["model_type"] == "full", [x for x in results_values.columns if "radii" in x]].to_numpy()
results_values = results_values[results_values["model_type"] == "feedback"]
results_values.loc[results_values["model_type"] == "feedback", "model_type"] = "feedback, trained separately"
results_all = pd.concat([results_estimates, results_values]).reset_index(drop=True)
feat_names = np.unique([x.split("r2_score_")[1].split("_num")[0] for x in results_all.columns if "r2_score_" in x])

results = pd.DataFrame()
num_eods_simulated = [int(x.split("num_eods_")[1]) for x in results_all.columns if "r2_score_" in x]
for num_eods in num_eods_simulated:
    temp_models = results_all[["rand_seed", "lambda_RC", "model_type"] + [f"r2_score_{feat_name}_num_eods_{num_eods}" for feat_name in feat_names]].copy()
    for feat_name in feat_names:
        temp_models[f"r2_score_{feat_name}"] = temp_models[f"r2_score_{feat_name}_num_eods_{num_eods}"]
        temp_models.drop(columns=[f"r2_score_{feat_name}_num_eods_{num_eods}"], inplace=True)
    temp_models["num_eods"] = num_eods
    results = pd.concat([results, temp_models]).reset_index(drop=True)
```

```python
temp_results = results[((results["lambda_RC"] == 1) | (results["lambda_RC"] == 2)) & (results["num_eods"] < 15)]
f, axes = plt.subplots(1, len(feat_names), figsize=(5 * len(feat_names), 4))
for i, (feat_n, plot_title) in enumerate(zip(["position_xs", "position_ys", "position_zs", "radii", "resistances", "capacitances"], ["Lateral position", "Distance", "Vertical position", "Radius", "Resistance", "Capacitance"])):
    axes[i].set_xticks([1,5,10])
    axes[i].set_ylim(0, 1.05)
    sns.lineplot(
        data=temp_results,
        x="num_eods",
        y=f"r2_score_{feat_n}",
        hue="model_type",
        ax=axes[i],
    )
    sns.scatterplot(
        data=temp_results,
        x="num_eods",
        y=f"r2_score_{feat_n}",
        hue="model_type",
        ax=axes[i],
        legend=False,
    )
    sns.move_legend(axes[i], "lower right", title="Model type")
    sns.despine(ax=axes[i], offset=10, trim=True)
    if i == 0:
        axes[i].set_ylabel(r"$R^2$ score")
    else:
        axes[i].set_ylabel("")
    axes[i].set_xlabel("Number of EODs")
    axes[i].set_title(plot_title)
plt.savefig("performance_summary.svg")
plt.show()
```
